```python
#filtering information to include only monthly data, creating the unemployment data frame
unemp_df = fred.search('unemployment rate state', filter=('frequency','Monthly'))
```

```python
#further filtering data to only include data in units percentages and are seasonally adjusted
unemp_df = unemp_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')
```

```python
#applying a string filter to only include rows with a title including the string 'Unemployment Rate'
unemp_df.loc[unemp_df['title'].str.contains('Unemployment Rate')]
```

```python
import time

all_results = []

for myid in unemp_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_results.append(results) #list to store all results, each element is its own data frame of state unemployments (index = year, column = state name)

#appling a sleep timer to slow down API commands, preventing a block
    time.sleep(0.1)
```

```python
#pulling participation rates

part_df = fred.search('participation rate state', filter=('frequency', 'Monthly'))
part_df = part_df.query('seasonal_adjustment == "Seasonally Adjusted" and units == "Percent"')

substring = 'Labor Force Participation Rate for'  #substring i am looking for

# filter rows where 'title' maches substring
part_df = part_df[part_df['title'].str.contains(substring, case=False)]



part_id_to_state = part_df['title'].str.replace('Labor Force Participation Rate for ', '').to_dict()

all_results = []

for myid in part_df.index:
  results = fred.get_series(myid)
  results = results.to_frame(name=myid)
  all_results.append(results)
  time.sleep(0.1)

part_states = pd.concat(all_results, axis=1)
part_states.columns = [part_id_to_state[c] for c in part_states.columns] #reusing dictionary to map participation rate state information with the state names
```

```python
#plotting unemployment rates and participation rates for each state
unemp_states = unemp_states.rename(columns={'the District of Columbia':'District Of Columbia'})
unemp_states.columns = part_states.columns

fig, axs = plt.subplots(10, 5, figsize=(30, 30), sharex=True)
axs = axs.flatten()

i = 0
for state in unemp_states.columns:
    if state in ["District Of Columbia","Puerto Rico"]:
        continue
    ax2 = axs[i].twinx() # creating another axis to fit borth unemp/participation rates
    unemp_states.query('index >= 2020 and index < 2022')[state].plot(ax=axs[i], label='Unemployment')
    part_states.query('index >= 2020 and index < 2022')[state].plot(ax=ax2, label='Participation', color=color_pal[1])
    ax2.grid(False)
    axs[i].set_title(state) #creating axis title
    i += 1
plt.tight_layout()
plt.show()
```
